```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv("sonar data.csv",header=None)
```

```python
sonar_data
sonar_data.head(10)
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

## R --> Rock And M --> Mine

```python
x=sonar_data.drop(columns=60)
y=sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,train_size=0.7,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Model

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

## Traning Acc

```python
x_train_pred=model.predict(x_train)
train_data_acc=accuracy_score(x_train_pred,y_train)
```

```python
print(train_data_acc)
```

## Testing Acc

```python
x_test_pred=model.predict(x_test)
test_data_acc=accuracy_score(x_test_pred,y_test)
```

```python
print(test_data_acc)
```
